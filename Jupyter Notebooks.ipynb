### Make and display charts!!!

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

df.plot.box()
````

````python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()

df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()

plt.figure()
df.plot();
````

## How can I install Jupyter?

First try

```
$ brew install jupyter
```

If that doesn't work out well, you can get it via Python's package management:

```
$ brew install python
$ pip3 install jupyter
```

Then run it using:

```
$ cd a-folder-where-I-want-to-store-my-notebooks
$ jupyter notebook
```

It'll then open up in your default browser.
